```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('courses.csv')
```

```python
df.head()
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
df.head()
```

```python
df
```

```python
def type(df):
    list_of_strings = df.split(' ')
    return list_of_strings[1]

df['type'] = df['modules_number'].apply(type)
```

```python
df
```

```python
def first_position_in_strings(df):
    list_of_strings = df.split(' ')
    return list_of_strings[0]

df['number_of_courses/modules'] = df['modules_number'].apply(first_position_in_strings)
df['level of course'] = df['level of course'].apply(first_position_in_strings)
```

```python
df
```

```python
df.drop('modules_number',axis=1,inplace = True)
```

```python
df
```

```python
df.info()
```

```python
def lowercase(df):
    return df.lower()
```

```python
df['level of course'] = df['level of course'].apply(lowercase)
df['title'] = df['title'].apply(lowercase)
```

```python
df = df[~df['rating'].isin(["What you'll learn", "Instructors",'Instructor'])]

df['rating'] = df['rating'].astype(float)
```

```python
df.info()
```

```python
df['number_of_courses/modules'].value_counts()
df = df[~df['number_of_courses/modules'].isin(['Guided'])]
```

```python
df['number_of_courses/modules'] = df['number_of_courses/modules'].astype(int)
```
